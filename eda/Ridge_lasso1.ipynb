# Data cleaning

```python
df[df.isnull().any(axis = 1)]
```

Up until 122: Bejaia Region

after 122: Sidi bel abbes region

so create a column and assign 0 to bejaia region and 1 to sidi bel abbes region

```python
df.loc[:122,"Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df
```

```python
df = df.dropna().reset_index(drop = True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df
```

```python
df.iloc[121:123,:]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
#changeing  data types to integers
df["day"] = df["day"].astype(int)
```

```python
df["month"] = df["month"].astype(int)
df["year"] = df["year"].astype(int)
df["Temperature"] = df["Temperature"].astype(int)
df["RH"] = df["RH"].astype(int)
df["Ws"] = df["Ws"].astype(int)
```

```python
df.info()
```

```python
df[["Rain","FFMC","DMC","DC","ISI","BUI","FWI"]] = df[["Rain","FFMC","DMC","DC","ISI","BUI","FWI"]].astype(float)
```
